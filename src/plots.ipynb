```python
import os
import json

import matplotlib.pyplot as plt
```

```python
dir_path = "/nlsasfs/home/isea/isea10/aansh/deception_probes/weights/Llama-3.1-8B-Instruct/Roleplaying/Poisoned_LTS_1"

accuracy_list = []
layer_list = range(32)
for i in range(32):
    file_path = os.path.join(dir_path, f"layer_{i}/eval_metrics.json")
    with open(file_path, "r") as f:
        data = json.load(f)
        accuracy_list.append(data["auroc"])


# plt.figure(figsize = (14, 8))    
# plt.plot(layer_list, accuracy_list, marker='o')
# plt.title('Layerwise Classification AUROC - Poisoned Roleplaying')
# plt.xlabel('Layer Number')
# plt.ylabel('AUROC')
# plt.xticks(layer_list)
```

```python
import pandas as pd
import seaborn as sns

# Create a Pandas DataFrame
data = pd.DataFrame({
    'Layer Number': layer_list,
    'AUROC': accuracy_list
})

# Set the black background style
plt.style.use('dark_background')

# Create the plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=data, x='Layer Number', y='AUROC', marker='o', color='cyan')

# Set titles and ticks
plt.title('Layerwise Classification AUROC - Poisoned Roleplaying - LTS-1', fontsize=16)
plt.xticks(layer_list)
plt.grid(axis='y', linestyle='--', alpha=0.6) # Optional: adds a grid for readability

plt.show()
```
